# Drop the "Method" Column

```python
dataset = dataset.fillna('Normal')
dataset.head()
```

```python
dataset.describe()
```

# List of the Participating Teams

```python
all_teams = dataset['Team1'].tolist() + dataset['Team2'].tolist()
all_teams = list(set(all_teams))
all_teams
```

# Number matches in each season

```python
matches_per_season = dataset.groupby('Season').ID.count()
matches_per_season
```

```python
plt.figure(figsize = (12,8))
plt.xticks(rotation = 75)
plt.title('Matches Per Season')
match_per_season_plot = sns.barplot(x = matches_per_season.index, y = matches_per_season)
match_per_season_plot.set(xlabel = 'Seasons', ylabel = 'No. of Matches');
```

# Total number of matches played by each team in each season

```python
total_matches_played = (dataset.Team1.value_counts() + dataset.Team2.value_counts()).sort_values(ascending = False)
total_matches_played
```

# The number of matches win by each team

```python
total_winning_count = dataset["WinningTeam"].value_counts()
total_winning_count
```

# Winning Percentage of each team

```python
win_percentage = (total_winning_count / total_matches_played).sort_values(ascending = False) * 100
win_percentage
```

```python
top10Groups = win_percentage.nlargest(10)
top10Groups
```

```python
plt.figure(figsize = (12,8))
plt.pie(top10Groups.values, labels =  top10Groups.index, autopct = '%.2f')
plt.title("Top 10 Teams Win Matches", fontsize = 20)
plt.plot()
```

# The number of matches each team has won in each season

```python
matches_won_each_season = dataset.groupby(["Season", "WinningTeam"]).size()
matches_won_each_season
```

```python
matches_won_each_season1 = pd.crosstab(dataset['WinningTeam'], dataset['Season'])
matches_won_each_season1
```

```python
plt.figure(figsize=(15,9))
plt.xlabel('Season', fontsize = 14)
plt.ylabel('Team Name', fontsize = 14)
plt.title('Number of matches each team has won in each season', fontsize = 20)
sns.heatmap(matches_won_each_season1, annot = True, cmap = 'flare')
```

# Teams selected for each season's finals

```python
ipl_final_teams = dataset[dataset['MatchNumber'] == 'Final']
ipl_final_teams
```

```python
ipl_winners = ipl_final_teams['WinningTeam'].value_counts()
ipl_winners
```

```python
plt.figure(figsize = (10,8))
plt.bar(ipl_winners.index, ipl_winners, color = 'red')
plt.xticks(fontsize = 12, rotation = 90)
plt.xlabel("Winning Team", fontsize = 16)
plt.ylabel("Winning Count", fontsize = 16)
plt.title("IPL Champions", fontsize = 20)
plt.tight_layout()
```
