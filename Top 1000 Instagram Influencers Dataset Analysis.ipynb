```python
dataset = dataset.drop('Country', axis=1)
```

```python
dataset.head()
```

```python
Audience_Country = dataset['Audience Country'].value_counts()
Audience_Country
```

```python
dataset['Audience Country'].value_counts(normalize = True).round(3)
```

```python
top10AudienceCountry = Audience_Country.nlargest(10)
top10AudienceCountry
```

```python
plt.figure(figsize = (12,8))
plt.pie(top10AudienceCountry.values, labels =  top10AudienceCountry.index, autopct = '%1.2f%%')
plt.title("Top 10 Audience Countries", fontsize = 20)
plt.plot()
```

```python
dataset = dataset.astype({'Followers' : "float"})
dataset
```

```python
dataset.info()
```

```python
top10followers = dataset.nlargest(10, 'Followers')
top10followers
```

```python
plt.figure(figsize = (12,8))
plt.bar(top10followers.Category, top10followers.Followers , color = 'blue')
plt.xticks(fontsize = 12, rotation = 90)
plt.xlabel("Category", fontsize = 16)
plt.ylabel("Followers in Million", fontsize = 16)
plt.title("Top 10 Categories by followers", fontsize = 20)
plt.tight_layout()
```
